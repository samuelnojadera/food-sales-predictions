# Data Analysis

Imports

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

Configs

```python
sns.set_theme()
```

Dataset

```python
df = pd.read_csv('sales_data.csv', index_col=0)
df.head()
```

```python
df.info()
```

### Understanding outlet information.

```python
unique_outlets = df.drop_duplicates(subset='Outlet_Identifier').sort_values(by='Outlet_Establishment_Year')
unique_years = unique_outlets['Outlet_Establishment_Year'].unique()
```

#### Outlets over time

```python
cumsum = []
for year in unique_years:
    filter = unique_outlets['Outlet_Establishment_Year']==year
    cumsum.append(len(unique_outlets[filter]))
cumsum = np.cumsum(cumsum)
outlets_overtime = sns.lineplot(x=unique_years, y=cumsum)
outlets_overtime.set_ylabel('Total Outlets')
outlets_overtime.set_xlabel('Year')
outlets_overtime.set_title('Outlets Over Time')
```

#### Outlet type, size, and location type distributions

```python
fig1, ax1 = plt.subplots(ncols=1, nrows=3)
fig1.tight_layout()
gb_outlet_type = unique_outlets.groupby('Outlet_Type').size()
gb_outlet_location_type = unique_outlets.groupby('Outlet_Location_Type').size()
gb_outlet_size = unique_outlets.groupby('Outlet_Size').size()
sns.barplot(y=gb_outlet_type.index, x=gb_outlet_type, ax=ax1[0])
sns.barplot(y=gb_outlet_location_type.index, x=gb_outlet_location_type, ax=ax1[1])
sns.barplot(y=gb_outlet_size.index, x=gb_outlet_size, ax=ax1[2])
for i in range(3):
    ax1[i].set_ylabel('')
```

### Item Properties and Distributions

```python
unique_items = df.drop_duplicates(subset='Item_Identifier')
```

#### Unique Items by type

```python
gb_unique_item_type = unique_items.groupby('Item_Type').size().sort_values(ascending=False)
ax_unique_item_type = sns.barplot(y=gb_unique_item_type.index, x=gb_unique_item_type)
ax_unique_item_type.set_title('Count of Unique Items by Type')
ax_unique_item_type.set_xlabel('Count')
ax_unique_item_type.set_ylabel('Item Type')
```

#### MRP grouped by Item Type

```python
ax_mrp_itemtype = sns.boxplot(x='Item_MRP', y='Item_Type', data=unique_items)
ax_mrp_itemtype.set_title('MRP by Item Type')
ax_mrp_itemtype.set_xlabel('MRP')
ax_mrp_itemtype.set_ylabel('Item Type')
```

### Reltionships

#### HeatMap of Item Visibility and Item Sales
Item_Outlet_Sales could be skewed by Item_MRP. So it's normalized by dividing it with its MRP.

```python
normalized_sales = df['Item_Outlet_Sales']/df['Item_MRP']
ax_visibility_sales = sns.histplot(x=df['Item_Visibility'], y=normalized_sales)
ax_visibility_sales.set_title('Visibility and Sales')
ax_visibility_sales.set_xlabel('Item Visibility')
ax_visibility_sales.set_ylabel('Item Sales')
```

#### Heatmap of Item MRP and ItemSales

```python
ax_sales_mrp = sns.histplot(y=df['Item_MRP'], x=normalized_sales)
ax_sales_mrp.set_title('Item MRP and Sales')
ax_sales_mrp.set_xlabel('Item Sales')
ax_sales_mrp.set_ylabel('MRP')
```
